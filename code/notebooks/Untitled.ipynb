```python
resolution = 0.25

for central_group in central_groups:
    plt.figure(figsize=(8, 4))
    plt.title("Cumulative datafraction " + central_group)
    
    for to_count_contact, contact_group in zip(to_count, contact_groups):
        
        if central_group == "RC6H5" and contact_group == "RC6H5":
            break
        
        result1 = "..\\results\\" + central_group + "\\" + central_group + "_" + contact_group + "_vdw.5\\"\
                  + central_group + "_" + contact_group + "_density.hdf"
        
        key = to_count_contact + str(resolution).rstrip("0").replace(".", "")
        df = pd.read_hdf(result1, key)
        
        df['datafrac_normalized'] = df[to_count_contact] / df[to_count_contact].sum()
        df = df.sort_values(by=to_count_contact, ascending=False)
        df['cumsum'] = df["datafrac_normalized"].cumsum()
        
        free_volume = 
#         

    # find the volume of the central group
#     tolerance = 0.5
#     contact_group_radius = get_vdw_distance_contact(df, settings)
#     volume = find_available_volume(avg_fragment=avg_frag, extra=(tolerance + contact_group_radius))
    
        plt.plot(df['cumsum'], range(0, len(df)), label=contact_group)
       
    plt.ylabel("Cumulative datafraction inside bins")
    plt.xlabel("Amount of bins")
#     plt.xlim(-100, 25000)
    plt.legend(loc="lower right")
    plt.show()

```
